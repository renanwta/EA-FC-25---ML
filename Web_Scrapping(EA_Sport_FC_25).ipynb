# Getting the stats

So to get the stats of our players we created a list for each stats that we want, after that we created a loop where we us the find_all to find all infomation with name 'tbody' with  the especific atributes and that we use a 'find' to find the 'div' with data-label with name of our stats, after that we save all the information in a one single list and than we transformed it in a DataSet

```python
list_grade_ovr = []
list_grade_pac = []
list_grade_sho = []
list_grade_pas = []
list_grade_dri = []
list_grade_def = []
list_grade_phy = []

url_loop = 'https://www.ea.com/en/games/ea-sports-fc/ratings?page='

for page in range (10):
  url_page = (url_loop+str(page))
  response = requests.get(url_page,headers=headers)
  site = BeautifulSoup(response.text, 'html.parser')

  for players in site.find_all('tbody',attrs = {'class':'Table_rowBlockWrapper__hA8k0'}):
    player_OVR = players.find('div',attrs = {'data-label':"OVR"})
    player_PAC = players.find('div',attrs = {'data-label':"PAC"})
    player_SHO = players.find('div',attrs = {'data-label':"SHO"})
    player_PAS = players.find('div',attrs = {'data-label':"PAS"})
    player_DRI = players.find('div',attrs = {'data-label':"DRI"})
    player_DEF = players.find('div',attrs = {'data-label':"DEF"})
    player_PHY = players.find('div',attrs = {'data-label':"PHY"})


    list_grade_ovr.append(player_OVR.text)
    list_grade_pac.append(player_PAC.text)
    list_grade_sho.append(player_SHO.text)
    list_grade_pas.append(player_PAS.text)
    list_grade_dri.append(player_DRI.text)
    list_grade_def.append(player_DEF.text)
    list_grade_phy.append(player_PHY.text)

list_grade = pd.concat([pd.Series(list_grade_ovr),pd.Series(list_grade_pac)],axis =1)
```

```python
df = pd.DataFrame({'OVR':list_grade_ovr,'PAC':list_grade_pac,
                   'SHO':list_grade_sho,'PAS':list_grade_pas,
                   'DRI':list_grade_dri,'DEF':list_grade_def,
                   'PHY':list_grade_phy})
df
```

# Data Threatment

On this part or data set need to pass for some steps to be ready for analysis

* The first steps is to remove the fist three letter at each cell;

* The second step is to remove the last number of when the cell have 3 numbers;

* And the final steps is to transform the type of the data

The first steps is to remove the 3 first letters, this letters indeicate the stats where and the number of the stats of each player, so to remove we used a lambda x to read each cell of our dataset and than asking to ignore the first three letter.

```python
for col in df.columns:
  df[col] = df[col].apply(lambda x:x[3:])
```

The second steps was the hardest step, because we was struggled think how we read each cell and than understand there is more number as expected and remove this number

So if we look to the type of the cell we can see that our values on the Dataset is a object, so it means that we can see each number as a separete letter, so have knowledge of this infomation we use a double loop to read each column and line like matrix with a conditional to see if the cell have more than 3 letter, if positive our we use a 'iloc' to bring to us only the 2 first letter.

```python
df.info()
```

```python
for i in range(df.shape[0]):
  for j in range(df.shape[1]):
    if len(df.iloc[i,j]) >= 3:
      df.iloc[i,j] = df.iloc[i,j][:2]
```

* The third and last part is to transform the type of our values, so we use the function of pandas 'pd.to_numeric' to transform our columns, and to do it on our dataset, we used a loop to read each columns and than make the type transformation.

```python
for col in df.columns:
  df[col] = pd.to_numeric(df[col])
```
